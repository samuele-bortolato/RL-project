```python
import os
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
RESULTS_FOLDER = 'runs_kaggle/Newtork/'
```

```python
SEEDS = []
for seed in os.scandir(RESULTS_FOLDER):
    path_to_folder_seed = next(os.scandir(next(os.scandir(seed.path)).path)).path
    path_to_results_file = os.path.join(path_to_folder_seed, 'results.pth')
    RUNS = torch.load(path_to_results_file)
    SEEDS.append(RUNS['RUNS'])

```

```python
plt.figure(figsize=(10,10))
color = plt.cm.rainbow(np.linspace(0, 1, len(SEEDS[0])))
for i, (seed0, seed1, seed2) in enumerate(zip(SEEDS[0], SEEDS[1], SEEDS[2])):
    
    plt.plot(seed0['VE'], color=color[i], alpha=0.3)
    plt.plot(seed1['VE'], color=color[i], alpha=0.3)
    plt.plot(seed2['VE'], color=color[i], alpha=0.3)
    plt.plot((seed0['VE']+seed1['VE']+seed2['VE'])/3, color=color[i], label=f"{seed0['input_type']}, {'symlog_output' if seed0['use_symlog'] else 'linear_output'}, {seed0['epochs']} epochs" )

plt.legend()
    
```

```python
color = plt.cm.rainbow(np.linspace(0, 1, len(SEEDS[0])))
for i, (seed0, seed1) in enumerate(zip(SEEDS[0], SEEDS[1],)):
    if seed0['use_symlog']:
        plt.plot(seed0['VE'], color=color[i], alpha=0.3)
        plt.plot(seed1['VE'], color=color[i], alpha=0.3)
        #plt.plot(seed2['VE'], color=color[i], alpha=0.3)
        plt.plot((seed0['VE']+seed1['VE'])/2, color=color[i], label=f"{seed0['input_type']}, {'symlog_output' if seed0['use_symlog'] else 'linear_output'}, {seed0['epochs']} epochs" )

plt.legend()
```

```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import pandas as pd
import os
from matplotlib import pyplot as plt
import matplotlib as mpl
mpl.style.use('default')
import numpy as np
kernel = np.ones(1000)
```

```python
log_dir='C:/Users/Sam/Downloads/results/runs/'
```

```python
R = {}
for experiment in os.listdir(log_dir):
    R[experiment]={}
    for run in os.listdir(os.path.join(log_dir,experiment)):
        R[experiment][run]={}
        for seed in os.listdir(os.path.join(log_dir,experiment,run)):

            file_name = os.listdir(os.path.join(log_dir,experiment,run,seed))[0]
            print(os.path.join(log_dir, experiment, run, seed, file_name).replace('\\','/'))

            acc = EventAccumulator(os.path.join(log_dir, experiment, run, seed, file_name).replace('\\','/'))
            acc.Reload()
            df = pd.DataFrame(acc.Scalars("Reward"))
            
            R[experiment][run][seed] = df
```

```python


E=R['PPO']

for run in E:
    for seed in E[run]:
        y = E[run][seed]['value'].to_numpy()
        plt.plot(np.convolve(y, kernel, 'same')/np.convolve(np.ones_like(y), kernel, 'same'), label=run)
plt.legend()
```
